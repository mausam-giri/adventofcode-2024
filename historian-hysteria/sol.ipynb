```python
left_list = []
right_list = []
with open("input.txt", "r") as f:
    for line in f:
        ll, rl = map(int, line.split())
        left_list.append(ll)
        right_list.append(rl)
```

```python
total_len = len(left_list)
total_len
```

```python
diff = 0
```

```python
left_list.sort()
right_list.sort()
for idx in range(len(left_list)):
    diff += abs(left_list[idx] - right_list[idx])

diff
```

## Part 2

```python
from collections import Counter
```

```python
right_count = Counter(right_list)
right_count
```

```python
sim_score = 0
```

```python
for item in left_list:
    sim_score += (item * right_count.get(item, 0))
```
